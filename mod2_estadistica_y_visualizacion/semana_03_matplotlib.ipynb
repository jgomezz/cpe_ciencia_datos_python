### Introducción

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-104, 105, 16)
y = x

plt.title("Función Lineal")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.plot(x,y)
plt.grid()
plt.show()

# print(x)
```

```python
x = np.arange(-104, 105, 16)
y = 4*x**2

def graficar(x, y, title):
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("F(x)")
    plt.plot(x,y)
    plt.grid()
    plt.show()

# Realizar gráfico

graficar(x,y,"Gráfico Parábola")
```

```python
x = np.arange(-104, 105, 16)
y = x**2 + 3*x + 5

graficar(x,y, "Ecuación Cuadrática")
```

### Subgráficos

```python

# Defino el rango de las x
x = np.arange(0, 2*np.pi, 0.05)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = 1/np.tan(x)

plt.subplot(2,2,1)
plt.title("Sin")
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.title("Cos")
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.title("Tan")
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.title("Cot")
plt.plot(x,y4)

plt.show()

```

## Tarea 3 
Realice un gráfico de línea de la función trigonométrica seno y coseno en dos gráficos distintos, cuyo dominio va de 0 a 2π, utilizando el método ‘subplot’.

```python
x = np.arange(0, 2*np.pi, 0.05)

y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1,2,1)
plt.title("Sin")
plt.plot(x,y1)
plt.grid()

plt.subplot(1,2,2)
plt.title("Cos")
plt.plot(x,y2)
plt.grid()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(0, 2*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1,2,1)
plt.title("Función Seno")
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.title("Función Coseno")
plt.plot(x,y2)

plt.show()
```

## Gráficos de Líneas


```python
x = np.arange(-10, 10, 0.1)

y1 = x
y2 = 4*x + 20
y3 = 8*x + 10

plt.title("Gráfico de líneas con Python")
plt.plot(x, y1, label ="línea y1")
plt.plot(x, y2, label ="línea y2")
plt.plot(x, y3, label ="línea y3")
plt.legend()

#plt.xlim(-5,5)
plt.savefig("output/fig-01")
plt.show()
```

```python
def mi_modelo(x):
    m = 3*pow(x,3) + 5*pow(x,2) + 4*x + 4
    m = np.log10(x) * m
    return m


# Defino el rango de las x
x = np.arange(0, np.pi/2, 0.05)

y1 = np.sin(x)
y2 = np.log10(x)
y3 = mi_modelo(x)

plt.title("Gráfico de líneas 2 con Python")
plt.plot(x, y1, label ="Sin")
plt.plot(x, y2, label ="Log")
plt.plot(x, y3, label ="Mi función")
plt.legend()

#plt.xlim(-5,5)
plt.xlabel("x")
plt.ylabel("y")

plt.savefig("output/fig-02")
plt.show()
```

## Gráficos de Barras

```python
import pandas as pd

nombre_archivo = "dataset/fmi/dataset_2025-11-22T23_33_05.997052849Z_DEFAULT_INTEGRATION_IMF.RES_WEO_9.0.0.csv"
df = pd.read_csv(nombre_archivo, thousands=",", decimal=".", sep=",")
df.set_index('COUNTRY', inplace=True)
df.head(2)
```

```python
anhos = list(map(str, range(2010, 2030)))

info_peru = df.loc["Peru", anhos]
info_peru.plot(kind="bar")
plt.title("GDP Perú : 2010-2029")
plt.xlabel("Años")
plt.ylabel("Miles de Millones $")
plt.grid()
plt.show()
```

```python
df.sort_values(by='2025', ascending=True, axis=0 , inplace=True)
df_top = df["2025"].tail(10)
df_top.plot(kind="barh")
plt.xlim(0,2500)

for index, value in enumerate(df_top):
    if not np.isnan(value):
        label = format(int(value), ',')
        plt.annotate(label, xy=(value,index-0.1))

plt.show()
```
