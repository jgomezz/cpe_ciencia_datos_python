## Descripción de las variables

- 'symboling' : Nivel de riesgo del seguro , de menor riesgo (-3) a mayor riesgto (3)
- "normalized-losses" : Es la perdida del pago promedio del seguro del auto por año y representa la perdida por año del auto. El valor oscila entre 65 a 256
- 'make' : Marca
- "aspiration" : Como el motor de combustión interno trabaja , puede ser estandar (std) o turbo 
- 'num-of-doors' : Número de puertas
- "body-style" : Carroceria
- 'drive-wheels' : Tracción de las ruedas
- 'engine-location' : Localización del motor
- 'wheel-base' : Distancia entre ejes 
- 'length' : Longitud
- 'width' : Ancho
- 'height': Alto
- 'curb-weight' : Peso del auto con equipación completa
- 'engine-type': Tipo de motor
- 'num-of-cylinders' : Número de cilindros
- 'engine-size' : Tamaño del motor
- 'fuel-system' : Sistema de combustible
- 'bore' : Diámetro
- 'stroke' : Distancia recorrida por el pistón durante cada ciclo
- 'compression-ratio': Ratio de comprensión
- 'horsepower' : Caballos de potencia
- 'horsepower-binned' : Caballos de potencia convertidos en una variable categórica 
- 'peak-rpm' : RPM pico 
- 'city-mpg' : consumo en la ciudad en millas por galon
- 'highway-mpg' : consumo en autopista en millas por galon
- 'price' : Precio
- 'city-L/100km' :  consumo en la ciudad en litros por kilometro
- 'diesel' : variable dummy para consumo por petróleo
- 'gas': variable dummy para consumo por gasoina

## Formateando datos


```python
import random

# Definir el rango de fechas
inicio = pd.to_datetime('2025-01-01 00:00:00')
fin = pd.to_datetime('2025-12-31 23:59:59')

# Función para generar una fecha y hora aleatoria dentro del rango
def fecha_hora_aleatoria(inicio, fin):
    delta = fin - inicio
    int_delta = int(delta.total_seconds())
    aleatorio_segundos = random.randint(0, int_delta)
    return inicio + pd.to_timedelta(aleatorio_segundos, unit='s')

# Generar la columna de fechas y horas aleatorias
df['date_time'] = [fecha_hora_aleatoria(inicio, fin) for _ in range(len(df))]
df.head()
```

```python
import datetime
df["date_time_new1"] = df["date_time"].apply(lambda x: x.strftime("%d-%m-%Y %H:%M:%S"))

df["date_time_new2"] = df["date_time"].apply(lambda x: x.strftime("%d-%m-%y %H:%M:%S"))

df["date_time_new3"] = df["date_time"].apply(lambda x: x.strftime("%m-%Y %H:%M:%S"))

df.head()
```

```python
df["day"] = df["date_time"].dt.day
df["day_week"] = df["date_time"].dt.day_name()
df["month"] = df["date_time"].dt.month
df["year"] = df["date_time"].dt.year
df["time"] = df["date_time"].dt.time
df.head()
```

### Normalización de datos

```python
# Obtenemos el mayor valor
max1 = df['length'].max()

# Obtenemos el menor valor
min1 = df['length'].min()

# Mostramos los datos
print("-----------------------------------------------------------")
print(df[['length']].head())

print("-----------------------------------------------------------")
# Reemplazamos (valor original - mínimo)/(máximo - mínimo)
df['length_norm1'] = (df['length'] - min1)/(max1 - min1)
df[['length_norm1']].head()
```

```python
# Obtenemos el mayor valor
max2 = df['width'].max()

# Obtenemos el menor valor
min2 = df['width'].min()

# Mostramos los datos
print("-----------------------------------------------------------")
print(df[['width']].head())

print("-----------------------------------------------------------")
# Reemplazamos (valor original - mínimo)/(máximo - mínimo)
df['width_norm1'] = (df['width']- min2)/(max2 - min2)
print(df[['width_norm1']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
A = scaler.fit_transform(df[["length"]])
B = scaler.fit_transform(df[["width"]])
df["length_norm2"] = pd.DataFrame(data=A)
df["width_norm2"] = pd.DataFrame(data=B)
print(df[['length_norm2','width_norm2']].head())
```

```python
# Convirtiendo mpg a km (0.4251 multiplicado por mpg)
print("-----------------------------------------------------------")
df['city-km'] = round(df["city-mpg"]*0.4251,3)
print(df[['city-mpg','city-km']].head())

print("-----------------------------------------------------------")
df['highway-km'] = round(df["highway-mpg"]*0.4251,3)
print(df[['highway-mpg','highway-km']].head())

print("-----------------------------------------------------------")
df.columns
```

## Variables numéricas y categóricas

### Estandarización de datos

```python
# Obtenemos la media
mean1 = df['length'].mean()
mean2 = df['width'].mean()

# Obtenemos la desviación estándar poblacional (ddof=0)
std1 = df['length'].std(ddof=0)
std2 = df['width'].std(ddof=0)

# Mostramos los datos
print("-----------------------------------------------------------")
print(df[['length','width']].head())

# Reemplazamos (valor original - media)/(desviación estándar)
df['length_std1'] = (df['length'] - mean1)/std1
df['width_std1'] = (df['width'] - mean2)/std2
print(df[['length_std1', 'width_std1']])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
A = scaler.fit_transform(df[["length"]])
B = scaler.fit_transform(df[["width"]])
df["length_std2"] = pd.DataFrame(data=A)
df["width_std2"] = pd.DataFrame(data=B)
print(df[['length_std2','width_std2']])
```
