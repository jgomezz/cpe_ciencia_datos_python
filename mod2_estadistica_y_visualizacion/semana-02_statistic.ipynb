# Estadísticos descriptivos con statsmodels

```python
import statsmodels.api as sm

a_matrix = np.arange(0,16)
a_matrix  

desc = sm.stats.DescrStatsW(a_matrix)

print(desc.mean)
print(desc.std)
print(desc.var)
print(desc.cov)
print(desc.quantile([0.25,0.5,0.75]))     
```

# Estadísticos descriptivos con scipy

```python
import scipy.stats as stats

a_matrix = np.arange(0,16)
a_matrix
```

```python
print("Minimo:", stats.tmin(a_matrix))
print("Maximo:", stats.tmax(a_matrix))
print("Media:", stats.tmean(a_matrix))
print("Varianza:", stats.tvar(a_matrix))
print("Desviacion Estandar:", stats.tstd(a_matrix)) 
print("Percentil 25:", stats.scoreatpercentile(a_matrix,25))
print("Percentil 50:", stats.scoreatpercentile(a_matrix,50))
print("Percentil 75:", stats.scoreatpercentile(a_matrix,75))
print("Rango intercuartilico:", stats.iqr(a_matrix))
print("Moda:", stats.mode(a_matrix))

```

# Estadísticos descriptivos para ndarray

```python
a_matrix = np.arange(0,16).reshape(4,4)
a_matrix
```

```python
print("Suma:", a_matrix.sum())
print("Media:", a_matrix.mean())
print("Minimo:", a_matrix.min())
print("Maximo:", a_matrix.max())
print("Varianza:", a_matrix.var())
print("Desviacion Estandar:", a_matrix.std()) 
```

```python
AXIS = 1 # 0 Columna , 1 : Fila
print(a_matrix)
print("Suma:", a_matrix.sum(AXIS))
print("Media:", a_matrix.mean(AXIS))
print("Minimo:", a_matrix.min(AXIS))
print("Maximo:", a_matrix.max(AXIS))
print("Varianza:", a_matrix.var(AXIS))
print("Desviacion Estandar:", a_matrix.std(AXIS)) 
```

# Estadísticos descriptivos con numpy

```python
a_matrix = np.arange(0,16).reshape(4,4)
a_matrix
```

```python
print("Sum:", np.sum(a_matrix))
print("Min:", np.min(a_matrix))
print("Max:", np.max(a_matrix))
print("Media:", np.mean(a_matrix))
print("Mediana:", np.median(a_matrix))
print("Moda:", stats.mode(a_matrix))
print("Varianza:", np.var(a_matrix))
print("Desviacion Estandar:", np.std(a_matrix))
print("Percentil 25:", np.percentile(a_matrix,25))
print("Percentil 50:", np.percentile(a_matrix,50))
print("Percentil 75:", np.percentile(a_matrix,75))


  
```

```python
# Estadísticos descriptivos con numpy axis=0 columnas, axis=1 filas
print("Sum:", np.sum(a_matrix, axis=0))
print("Min:", np.min(a_matrix, axis=0))
print("Max:", np.max(a_matrix, axis=0))
print("Media:", np.mean(a_matrix, axis=0))
print("Mediana:", np.median(a_matrix, axis=0))
print("Moda:", stats.mode(a_matrix, axis=0))
print("Varianza:", np.var(a_matrix, axis=0))
print("Desviacion Estandar:", np.std(a_matrix,axis=0))
print("Percentil 25:", np.percentile(a_matrix,25,axis=0))
print("Percentil 50:", np.percentile(a_matrix,50,axis=0))
print("Percentil 75:", np.percentile(a_matrix,75,axis=0))
```

# Estadísticos descriptivos con Pandas
