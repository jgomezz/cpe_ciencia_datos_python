### Introducción

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-104, 105, 16)
y = x

plt.title("Función Lineal")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.plot(x,y)
plt.grid()
plt.show()

# print(x)
```

```python
x = np.arange(-104, 105, 16)
y = 4*x**2

def graficar(x, y, title):
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("F(x)")
    plt.plot(x,y)
    plt.grid()
    plt.show()

# Realizar gráfico

graficar(x,y,"Gráfico Parábola")
```

```python
x = np.arange(-104, 105, 16)
y = x**2 + 3*x + 5

graficar(x,y, "Ecuación Cuadrática")
```
