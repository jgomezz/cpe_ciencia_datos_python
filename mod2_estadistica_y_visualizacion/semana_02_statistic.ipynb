# Estadísticos descriptivos con statsmodels

- https://www.statsmodels.org/stable/index.html

```python
import statsmodels.api as sm

a_matrix = np.arange(0,16)
print(a_matrix)
```

```python
desc = sm.stats.DescrStatsW(a_matrix)

print(desc.mean)
print(desc.std)
print(desc.var)
print(desc.cov)
print(desc.quantile([0.25,0.5,0.75]))    
```

# Estadísticos descriptivos con scipy
- https://scipy.org/

```python
import scipy.stats as stats

a_matrix = np.arange(0,16)
a_matrix
```

```python
print("Minimo:", stats.tmin(a_matrix))
print("Maximo:", stats.tmax(a_matrix))
print("Media:", stats.tmean(a_matrix))
print("Varianza:", stats.tvar(a_matrix))
print("Desviacion Estandar:", stats.tstd(a_matrix)) 
print("Percentil 25:", stats.scoreatpercentile(a_matrix,25))
print("Percentil 50:", stats.scoreatpercentile(a_matrix,50))
print("Percentil 75:", stats.scoreatpercentile(a_matrix,75))
print("Rango intercuartilico:", stats.iqr(a_matrix))
print("Moda:", stats.mode(a_matrix))

```

### Ejercicio 2 : Cree una función para calcular la varianza poblacional y otra para calcular la varianza muestral que admita como argumento un ndarray. Aplíquelo al siguiente conjunto de edades: 22, 18, 35, 23, 41, 28, 45, y compare los resultados con los obtenidos mediante los métodos de numpy para calcular la varianza poblacional y muestral.

```python
def varianza_poblacional(datos):
    media = np.mean(datos)
    return np.sum((datos - media)**2) / len(datos)

def varianza_muestral(datos):
    media = np.mean(datos)
    return np.sum((datos - media)**2) / (len(datos) - 1)

edades = np.array([22, 18, 35, 23, 41, 28, 45])

print(varianza_poblacional(edades))
print(varianza_muestral(edades))

print(edades)

print("NumPy poblacional:", np.var(edades))
print("NumPy muestral:", np.var(edades, ddof=1))
```

# Cargar CSV y asignar nombres a columnas

Data set  : https://archive.ics.uci.edu/dataset/53/iris

```python
import pandas as pd
import numpy as np


data = \
    pd.read_csv("iris.csv", 
        names=["sepal_length", "sepal_width", \
             "petal_length", "petal_width", "class"])
data
```

```python
media = data["sepal_length"].mean()

#print(round(media,2))

nro = len(data["sepal_length"]) 
nro
```

```python
# Calcular el intervalo de confianza al 95% para la media de la longitud del sépalo.

import scipy.stats as st

nro = len(data["sepal_length"])
media = data["sepal_length"].mean()

# Como tengo mas de 30 muestras uso la distribución normal para calcular el intervalo de confianza
ic = st.norm.interval(confidence = 0.95, loc=media, scale=st.sem(data["sepal_length"]))

print("La media de la longitud del sépalo:", media)
print("Intervalo de confianza al 95% para la media de la longitud del sépalo:", ic)
```
