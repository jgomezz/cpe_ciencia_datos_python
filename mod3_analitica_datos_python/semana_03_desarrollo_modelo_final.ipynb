## Regresión múltiple

#### Ecuación de regresión múltiple

$ y = b_0 + b_1x_1 + b_2x_2 +...+ b_nx_n + error $

$ \hat{y} = b_0 + b_1x_1 + b_2x_2 +...+ b_nx_n $

$ y = \hat{y} + error $

$ y - \hat{y} = error $

```python
cdf = df[['horsepower','highway-mpg','compression-ratio','price']]
cdf.head()
```

```python
# Regresión usando la librería sklearn
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
x = cdf[['horsepower','highway-mpg','compression-ratio']]
y = cdf[['price']]
```

```python
lm.fit(x, y)
b0 = lm.intercept_
b = lm.coef_
print('Intercepto =', b0)
print('Coeficientes =', b)
```

$$ \hat{y} = 569.511 + 148.563x_1 - 272.354x2 + 562.365x_3 $$

## Regresión polinómica

<center><b>Cuadrática (segundo orden)</b></center>
$$
\hat{y} = b_0 + b_1x + b_2x^2\\
$$

<center><b>Cúbica (tercer orden)</b></center>
$$
\hat{y} = b_0 + b_1x + b_2x^2 + b_3x^3\\
$$

<center><b>Superior (n-ésimo orden)</b>:</center>
$$
\hat{y} = b_0 + b_1x + b_2x^2 + b_3x^3 +...+ b_nx^n\\
$$

```python
# Regresión polinómica de grado 2
x = df['highway-mpg']
y = df['price']
grado = 2
f = np.polyfit(x, y, grado)
print("Coeficientes =" ,f)
```

```python
p = np.poly1d(f)
print(p)
```

$$ \hat{y} = 86265.833 - 3870.871x + 46.452x^2 $$

```python
def plot_poly(model, independent_variable, dependent_variable, name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)
    plt.plot(independent_variable, dependent_variable, '.', x_new, y_new, '-')
    plt.title('Gráfica de regresión polinómica de grado 2')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(name)
    plt.ylabel('price')
    plt.show()

plot_poly(p, x, y, 'highway-mpg')
```

## Predicciones

```python
df1 = df[['horsepower','price']]
df1
```

```python
# Regresión usando la librería statsmodels
import statsmodels.formula.api as smf
results = smf.ols('price ~  horsepower', data=df1).fit()
print(results.summary())
```

```python
y_pred = results.predict(df1['horsepower'])
df1["price_hat"] = y_pred
df1["error"] = df1["price"] - df1["price_hat"] 
df1
```

```python
df2 = df[['highway-mpg','price']]
df2 = df2.rename(columns={'highway-mpg': 'highway_mpg'})
df2
```

```python
# Regresión usando la librería statsmodels
import statsmodels.formula.api as smf
results = smf.ols('price ~  highway_mpg', data=df2).fit()
print(results.summary())
```

```python
y_pred = results.predict(df2['highway_mpg'])
df2["price_hat"] = y_pred
df2["error"] = df2["price"] - df2["price_hat"] 
df2
```

```python
df4 = df[['horsepower','highway-mpg','compression-ratio','price']]
df4 = df4.rename(columns={'highway-mpg': 'highway_mpg',
                          'compression-ratio': 'compression_ratio'})
df4
```

```python
# Regresión usando la librería statsmodels
import statsmodels.formula.api as smf
results = smf.ols('price ~  horsepower + highway_mpg + compression_ratio', data=df4).fit()
print(results.summary())
```

```python
y_pred = results.predict(df4[['horsepower','highway_mpg','compression_ratio']])
df4["price_hat"] = y_pred
df4["error"] = df4["price"] - df4["price_hat"] 
df4
```
