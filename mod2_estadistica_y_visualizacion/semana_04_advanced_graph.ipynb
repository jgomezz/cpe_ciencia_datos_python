### Tarea 1

Hacer un gráfico de pie con las PBI del año 2025 de los paises de Peru, Chile, Argentina, Colombia, Brasil y Mexico

```python
## Lectura de datos desde un archivo CSV

import pandas as pd

filename = "dataset/fmi/WEO_Data_Latinoamerica_2025.csv"

df = pd.read_csv(filename)

#df.head()

df.set_index('COUNTRY', inplace=True)

df_filter = df.loc[["Peru", "Chile", "Argentina", "Colombia", "Brazil" ,"Mexico"],'2025']

df_filter.head()
```

```python
import pandas as pd

data = {'Matemática': [10,15,20,15,5,5,5,10,19,19,19],
        'Lenguaje': [3,5,7,6,7,8,5,7,10,8,19],
        'Historia': [19,18,17,12,17,18,15,3,10,18,19]}
df_subjects = pd.DataFrame(data)
#df.head()

df_subjects.plot(kind="box")
plt.title("Notas de 2ro Secundaria")
plt.ylabel("Rango de notas")
plt.savefig("output/boxplot-04")
plt.show()
```

```python
df_subjects.describe()
```

```python
# Obtener los outliers
def get_outliers(df,colname):
    Q1 = df.describe()[colname]['25%']
    Q3 = df.describe()[colname]['75%']
    IQ = Q3 - Q1
    upper_threshold =  Q3 + 1.5*IQ
    lower_threshold =  Q1 - 1.5*IQ
    outliers = df.loc[ (df[colname] < lower_threshold) | ( df[colname] > upper_threshold), colname]
    return outliers 

#print(type(get_outliers(df,"Matemática")))

print("Outlier Matemática -->")
for id, value in get_outliers(df_subjects,"Matemática").items():
    print(id, value)
print("----------------------------------------------")
print("Outlier Lenguaje -->")
for id, value in get_outliers(df_subjects,"Lenguaje").items():
    print(value)
print("----------------------------------------------")
print("Outlier Historia -->")
for id, value in get_outliers(df_subjects,"Historia").items():
    print(value) 
```

#### Diagrama de caja a partir de un archivo CSV


```python
## Lectura de datos desde un archivo CSV

import pandas as pd

filename = "dataset/fmi/WEO_Data_Latinoamerica_2025.csv"
df = pd.read_csv(filename)
df.set_index('COUNTRY', inplace=True)
list_country = ["Peru", "Chile", "Argentina", "Colombia", "Brazil" ,"Mexico"]
range_year = [str(x) for x in range(2020, 2030)]
df_countries = df.loc[list_country,range_year]
df_countries = df_countries.T
df_countries.head()
```

```python
df_countries.plot(kind="box")
```

```python
for index, value in get_outliers(df_countries,"Argentina").items():
    print(index,"-->", value)
```
