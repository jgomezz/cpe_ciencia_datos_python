```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-104, 105, 16)
y = x

plt.title("Función Lineal")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.plot(x,y)
plt.grid()
plt.show()

# print(x)
```
