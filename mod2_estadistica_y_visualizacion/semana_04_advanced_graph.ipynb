### Global Demand Smartphone : 3Q-2025 Shipment Volumes

Source : https://my.idc.com/getdoc.jsp?containerId=prUS53868725

<img src="imagenes/GDS_3Q_2025.png" width="800"/>

```python
# Cantidad de móviles vendidos en Q3 - 2025
etiquetas = ['Samsung', 'Apple', 'Xiaomi', 'Transsion', 'vivo', 'Otros']
valores = [61.4, 58.6, 43.5, 29.2, 28.8, 101.2]
colores = ['green', 'red', 'blue', 'purple', 'yellow', 'orange']

plt.title("Participación en la venta mundial de móviles del Q3 del 2025")
plt.pie(valores, labels=etiquetas, colors=colores)
plt.savefig("output/pie-01")
plt.show()
```

```python
# Cantidad de móviles vendidos en Q3 - 2025 - Girado
etiquetas = ['Samsung', 'Apple', 'Xiaomi', 'Transsion', 'vivo', 'Otros']
valores = [61.4, 58.6, 43.5, 29.2, 28.8, 101.2]
colores = ['green', 'red', 'blue', 'purple', 'yellow', 'orange']

plt.title("Participación en la venta mundial de móviles del Q3 del 2025 - Girado")
plt.pie(valores, labels=etiquetas, colors=colores, startangle=90) # Gira 90 grados
plt.savefig("output/pie-02")
plt.show()
```

```python
# Cantidad de móviles vendidos en Q3 - 2025 - Tajadas
etiquetas = ['Samsung', 'Apple', 'Xiaomi', 'Transsion', 'vivo', 'Otros']
valores = [61.4, 58.6, 43.5, 29.2, 28.8, 101.2]
colores = ['green', 'red', 'blue', 'purple', 'yellow', 'orange']
explode = [0.1, 0, 0.5, 0, 0, 0]
plt.title("Participación en la venta mundial de móviles del Q3 del 2025 - Tajadas")
plt.pie(valores, labels=etiquetas, colors=colores, explode=explode) # Tajadas
plt.savefig("output/pie-03")
plt.show()
```

```python
font = {'family': 'serif',
        'color': 'green',
        'weight': 'normal',
        'size': 20,
        }

# Cantidad de móviles vendidos en Q3 - 2025
etiquetas = ['Samsung', 'Apple', 'Xiaomi', 'Transsion', 'vivo', 'Otros']
valores = [61.4, 58.6, 43.5, 29.2, 28.8, 101.2]
colores = ['green', 'red', 'blue', 'purple', 'yellow', 'orange']

plt.title("Participación en la venta mundial de móviles del Q3 del 2025", fontdict=font)
plt.pie(valores, labels=etiquetas, colors=colores, shadow=True)
plt.savefig("output/pie-01")
plt.show()
```

```python
font = {'family': 'serif',
        'color': 'green',
        'weight': 'normal',
        'size': 20,
        }

# Cantidad de móviles vendidos en Q3 - 2025
etiquetas = ['Samsung', 'Apple', 'Xiaomi', 'Transsion', 'vivo', 'Otros']
valores = [61.4, 58.6, 43.5, 29.2, 28.8, 101.2]
colores = ['green', 'red', 'blue', 'purple', 'yellow', 'orange']

plt.title("Participación en la venta mundial de móviles del Q3 del 2025", fontdict=font)
plt.pie(valores, labels=etiquetas, colors=colores, shadow=True, autopct='%1.1f%%', textprops={"fontsize" :16})
plt.savefig("output/pie-01")
plt.show()
```
