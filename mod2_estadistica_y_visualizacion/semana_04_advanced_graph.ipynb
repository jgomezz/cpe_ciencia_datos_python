```python
# Obtener los outliers
def get_outliers(df,colname):
    """
    Obtiene los outliers de una columna específica de un DataFrame utilizando el método del rango intercuartílico (IQR).
    """ 
    Q1 = df.describe()[colname]['25%']
    Q3 = df.describe()[colname]['75%']
    IQ = Q3 - Q1
    upper_threshold =  Q3 + 1.5*IQ
    lower_threshold =  Q1 - 1.5*IQ
    #outliers = df.loc[ (df[colname] < lower_threshold) | ( df[colname] > upper_threshold), colname]
    outliers = df.loc[ (df[colname] < lower_threshold) | ( df[colname] > upper_threshold), colname]
    return outliers 

#print(type(get_outliers(df,"Matemática")))
```

```python
print("Outlier Matemática -->")
for id, value in get_outliers(df_subjects,"Matemática").items():
    print(id, value)
```

```python
print("----------------------------------------------")
print("Outlier Lenguaje -->")
for id, value in get_outliers(df_subjects,"Lenguaje").items():
    print(value)
```

```python
import requests

def get_imf_data(indicador, pais, años=None):
    url = f"https://www.imf.org/external/datamapper/api/v2/{indicador}/{pais}"
    data = requests.get(url).json()
    
    serie = data["values"][indicador][pais]
    
    if años:
        años_str = [str(a) for a in años]
        serie = {a: v for a, v in serie.items() if a in años_str}
    
    return serie

# Uso
pib = get_imf_data("NGDPD", "PER", años=range(2019, 2033))

for año, valor in sorted(pib.items()):
    print(f"{año}: US$ {valor:,.2f} mil millones")
```
