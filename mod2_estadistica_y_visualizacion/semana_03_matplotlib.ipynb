### Introducción

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-104, 105, 16)
y = x

plt.title("Función Lineal")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.plot(x,y)
plt.grid()
plt.show()

# print(x)
```

```python
x = np.arange(-104, 105, 16)
y = 4*x**2

def graficar(x, y, title):
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("F(x)")
    plt.plot(x,y)
    plt.grid()
    plt.show()

# Realizar gráfico

graficar(x,y,"Gráfico Parábola")
```

```python
x = np.arange(-104, 105, 16)
y = x**2 + 3*x + 5

graficar(x,y, "Ecuación Cuadrática")
```

### Subgráficos

```python

# Defino el rango de las x
x = np.arange(0, 2*np.pi, 0.05)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = 1/np.tan(x)

plt.subplot(2,2,1)
plt.title("Sin")
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.title("Cos")
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.title("Tan")
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.title("Cot")
plt.plot(x,y4)

plt.show()

```

## Gráficos de Líneas


```python
x = np.arange(-10, 10, 0.1)

y1 = x
y2 = 4*x + 20
y3 = 8*x + 10

plt.title("Gráfico de líneas con Python")
plt.plot(x, y1, label ="línea y1")
plt.plot(x, y2, label ="línea y2")
plt.plot(x, y3, label ="línea y3")
plt.legend()

#plt.xlim(-5,5)
plt.savefig("fig-01")
plt.show()
```

```python
def mi_modelo(x):
    m = 3*pow(x,3) + 5*pow(x,2) + 4*x + 4
    m = np.log10(x) * m
    return m


# Defino el rango de las x
x = np.arange(0, np.pi/2, 0.05)

y1 = np.sin(x)
y2 = np.log10(x)
y3 = mi_modelo(x)

plt.title("Gráfico de líneas 2 con Python")
plt.plot(x, y1, label ="Sin")
plt.plot(x, y2, label ="Log")
plt.plot(x, y3, label ="Mi función")
plt.legend()

#plt.xlim(-5,5)
plt.xlabel("x")
plt.ylabel("y")

plt.savefig("fig-02")
plt.show()
```
