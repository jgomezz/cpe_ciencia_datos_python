## Descripción de las variables

- 'symboling' : Nivel de riesgo del seguro , de menor riesgo (-3) a mayor riesgto (3)
- "normalized-losses" : Es la perdida del pago promedio del seguro del auto por año y representa la perdida por año del auto. El valor oscila entre 65 a 256
- 'make' : Marca
- "aspiration" : Como el motor de combustión interno trabaja , puede ser estandar (std) o turbo 
- 'num-of-doors' : Número de puertas
- "body-style" : Carroceria
- 'drive-wheels' : Tracción de las ruedas
- 'engine-location' : Localización del motor
- 'wheel-base' : Distancia entre ejes 
- 'length' : Longitud
- 'width' : Ancho
- 'height': Alto
- 'curb-weight' : Peso del auto con equipación completa
- 'engine-type': Tipo de motor
- 'num-of-cylinders' : Número de cilindros
- 'engine-size' : Tamaño del motor
- 'fuel-system' : Sistema de combustible
- 'bore' : Diámetro
- 'stroke' : Distancia recorrida por el pistón durante cada ciclo
- 'compression-ratio': Ratio de comprensión
- 'horsepower' : Caballos de potencia
- 'horsepower-binned' : Caballos de potencia convertidos en una variable categórica 
- 'peak-rpm' : RPM pico 
- 'city-mpg' : consumo en la ciudad en millas por galon
- 'highway-mpg' : consumo en autopista en millas por galon
- 'price' : Precio
- 'city-L/100km' :  consumo en la ciudad en litros por kilometro
- 'diesel' : variable dummy para consumo por petróleo
- 'gas': variable dummy para consumo por gasoina

```python
# Imputación para la variable bore
avg_bore = df['bore'].astype('float').mean()
avg_bore = round(avg_bore,4)      
print("El promedio de la variable bore es", avg_bore)
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
# Imputación para la variable normalized-losses
avg_norm_loss = df["normalized-losses"].astype("float").mean()
avg_norm_loss = round(avg_norm_loss,4)    
print("El promedio de normalized-losses es", avg_norm_loss)

# df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
# Usar fillna en lugar de replace
df["normalized-losses"] = df["normalized-losses"].fillna(avg_norm_loss)
```

```python
# Imputación para la variable horsepower
avg_horsepower=df['horsepower'].astype('float').mean()
avg_horsepower = round(avg_horsepower,4)
print("El promedio de la variable horsepower es", avg_horsepower)
df["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
```

```python
# Imputación para la variable peak-rpm
avg_peak_rpm=df['peak-rpm'].astype('float').mean()
avg_peak_rpm = round(avg_peak_rpm,4)
print("El promedio de la variable peak-rpm es", avg_peak_rpm)
df["peak-rpm"].replace(np.nan, avg_peak_rpm, inplace=True)
```

```python
# Promedio para un grupo de variables
lista = ["normalized-losses","stroke","horsepower","peak-rpm"]
for column in lista:
    avg = df[column].astype('float').mean(axis=0)
    avg = round(avg,4)
    df[column].replace(np.nan, avg, inplace=True)
    print("El promedio de la variable" ,column,"es", avg)
```

```python
# Imputación de las variables stroke, peak-rpm, horsepower y normalized-losses
for i in ["stroke","peak-rpm","horsepower","normalized-losses"]:
    avg = df[i].astype("float").mean()
    df[i].replace(np.nan, avg, inplace=True)
    
df[["stroke","peak-rpm","horsepower","normalized-losses"]]
```

```python
import random

# Definir el rango de fechas
inicio = pd.to_datetime('2024-01-01 00:00:00')
fin = pd.to_datetime('2024-12-31 23:59:59')

# Función para generar una fecha y hora aleatoria dentro del rango
def fecha_hora_aleatoria(inicio, fin):
    delta = fin - inicio
    int_delta = int(delta.total_seconds())
    aleatorio_segundos = random.randint(0, int_delta)
    return inicio + pd.to_timedelta(aleatorio_segundos, unit='s')

# Generar la columna de fechas y horas aleatorias
df['date_time'] = [fecha_hora_aleatoria(inicio, fin) for _ in range(len(df))]
print(df)
```

```python
import datetime
df["date_time_new1"] = df["date_time"].\
                      apply(lambda x: x.strftime("%d-%m-%Y %H:%M:%S"))

df["date_time_new2"] = df["date_time"].\
                      apply(lambda x: x.strftime("%d-%m-%y %H:%M:%S"))

df["date_time_new3"] = df["date_time"].\
                      apply(lambda x: x.strftime("%m-%Y %H:%M:%S"))
print(df)
```

```python
df["day"] = df["date_time"].dt.day
df["day_week"] = df["date_time"].dt.day_name()
df["month"] = df["date_time"].dt.month
df["year"] = df["date_time"].dt.year
df["time"] = df["date_time"].dt.time
print(df)
```

### Normalización de datos

```python
# Obtenemos el mayor valor
max1 = df['length'].max()
max2 = df['width'].max()

# Obtenemos el menor valor
print("-----------------------------------------------------------")
min1 = df['length'].min()
min2 = df['width'].min()

# Mostramos los datos
print("-----------------------------------------------------------")
print(df[['length','width']].head())

# Reemplazamos (valor original - mínimo)/(máximo - mínimo)
df['length_norm1'] = (df['length'] - min1)/(max1 - min1)
df['width_norm1'] = (df['width']- min2)/(max2 - min2)
print(df[['length_norm1','width_norm1']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
A = scaler.fit_transform(df[["length"]])
B = scaler.fit_transform(df[["width"]])
df["length_norm2"] = pd.DataFrame(data=A)
df["width_norm2"] = pd.DataFrame(data=B)
print(df[['length_norm2','width_norm2']].head())
```

```python
# Convirtiendo mpg a km (0.4251 multiplicado por mpg)
print("-----------------------------------------------------------")
df['city-km'] = round(df["city-mpg"]*0.4251,3)
print(df[['city-mpg','city-km']].head())

print("-----------------------------------------------------------")
df['highway-km'] = round(df["highway-mpg"]*0.4251,3)
print(df[['highway-mpg','highway-km']].head())

print("-----------------------------------------------------------")
df.columns
```

## Variables numéricas y categóricas

### Estandarización de datos

```python
# Obtenemos la media
mean1 = df['length'].mean()
mean2 = df['width'].mean()

# Obtenemos la desviación estándar poblacional (ddof=0)
std1 = df['length'].std(ddof=0)
std2 = df['width'].std(ddof=0)

# Mostramos los datos
print("-----------------------------------------------------------")
print(df[['length','width']].head())

# Reemplazamos (valor original - media)/(desviación estándar)
df['length_std1'] = (df['length'] - mean1)/std1
df['width_std1'] = (df['width'] - mean2)/std2
print(df[['length_std1', 'width_std1']])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
A = scaler.fit_transform(df[["length"]])
B = scaler.fit_transform(df[["width"]])
df["length_std2"] = pd.DataFrame(data=A)
df["width_std2"] = pd.DataFrame(data=B)
print(df[['length_std2','width_std2']])
```

```python
# Ver tipo de variable
print(df[["horsepower"]].dtypes)

# Se convierte a enteros
print("-----------------------------------------------------------")
df["horsepower"] = df["horsepower"].astype(int)
print(df[["horsepower"]].dtypes)

# Graficando histograma para horsepower
print("-----------------------------------------------------------")
import matplotlib.pyplot as plt

plt.hist(df["horsepower"]) # El valor por defecto bins = 10
plt.xlabel("horsepower")
plt.ylabel("Número de veces")
plt.title("horsepower bins")
```

```python
# Generar puntos de corte 
minimo = df["horsepower"].min()
maximo = df["horsepower"].max()
punto_de_corte = 4
bins = np.linspace(minimo,maximo,punto_de_corte)
print("Particiones =",bins)

print("-----------------------------------------------------------")
bins_custom = [48, 100, 200, 262] # cortes personalizados
group_names = ['Bajo', 'Medio', 'Alto']

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names)
df['horsepower-binned-custom'] = pd.cut(df['horsepower'], bins_custom, labels=group_names)
print(df[['horsepower','horsepower-binned','horsepower-binned-custom']].head())
```
